```python
#Rename columns with suitable name
rename_columns =Crickets_data.rename(columns={'BF': 'Balls_Faced', 'SR': 'Batting_Strike_Rate', 'HS': 'Highest_Inns_Score', 'NO': 'Not_Outs', 'Mat': 'Matches'})
```

```python
### Split the 'Span' column into two columns (Start and End Date).
rename_columns[['Start_Date', 'End_Date']] = rename_columns['Span'].str.split('-', expand=True)
rename_columns.head()
```

```python
rename_columns= rename_columns.drop(columns=['Span'])
```

```python
#change type of columns from str to int =>start_date, end_date
rename_columns['Start_Date'] = rename_columns['Start_Date'].astype(int)
rename_columns['End_Date'] = rename_columns['End_Date'].astype(int)
```

```python
rename_columns.dtypes
```

```python
### Split up the country from the 'Player' column.
rename_columns['Country'] = rename_columns['Player'].str.split('(', expand=True)[1].str.replace(')', '')
rename_columns.head(2)
```

```python
### Delete the country from the 'Player' column.
rename_columns['Player'] = rename_columns['Player'].replace(r"\s*\(.*\)", "", regex=True)
rename_columns.head()
```

```python
### Fixing the Data Types of the columns.

rename_columns.dtypes
```

```python
rename_columns.columns
```

```python
rename_columns['Highest_Inns_Score'].unique()
```

#### We can find that few values are ending with '*'s. So we can remove those '*'s and convert the data type into 'int'.

```python
rename_columns['Highest_Inns_Score'] = rename_columns['Highest_Inns_Score'].replace(r'\*', '', regex=True)
rename_columns['Highest_Inns_Score'] = rename_columns['Highest_Inns_Score'].astype(int)
```

```python
rename_columns['Balls_Faced'].unique()
```

```python
rename_columns['Balls_Faced'] = rename_columns['Balls_Faced'].replace(r'\+','',regex=True)
rename_columns['Balls_Faced'] = rename_columns['Balls_Faced'].replace(r'\-',0,regex=True)
rename_columns['Balls_Faced'] = rename_columns['Balls_Faced'].astype(int)
```

```python
#4s','6s'
rename_columns['4s'].unique()
```

```python
rename_columns['4s']= rename_columns['4s'].replace(r"\+","",regex=True)
rename_columns['4s']= rename_columns['4s'].astype(int)
```

```python
rename_columns['6s'].unique()
```

```python
rename_columns['6s']= rename_columns['6s'].replace(r"\+","",regex=True)
rename_columns['6s']= rename_columns['6s'].astype(int)
```

```python
rename_columns['Batting_Strike_Rate'].unique()
```

```python
rename_columns['Batting_Strike_Rate'] = rename_columns['Batting_Strike_Rate'].replace(r'\-', 0.0, regex=True)
rename_columns['Batting_Strike_Rate'] = rename_columns['Batting_Strike_Rate'].replace(r'', 0, regex=True)
rename_columns['Batting_Strike_Rate'] = rename_columns['Batting_Strike_Rate'].astype(float)
```

```python
rename_columns['0'].unique()
```

```python
rename_columns['0'] = rename_columns['0'].replace(r'\-', 0, regex=True)
rename_columns['0'] = rename_columns['0'].astype(int)
```

```python
rename_columns.dtypes
```

```python
clear_data = rename_columns.copy()
```

### Add a 'Career_Length' column based on the 'Star_Date' and 'End_Date' columns.

```python
clear_data['Career_Length'] = clear_data['End_Date'] - clear_data['Start_Date']
```

```python
clear_data['Career_Length'].unique()
```

```python
clear_data.dtypes
```

### First calculations to Perform on the DataFrame:
1. Calculate the average career length of cricketers.
2. Compute the average batting strike rate for cricketers who played for more than 10 years.
3. Determine the number of cricketers who played before 1960.
4. Find the maximum highest innings score by country.
5. Calculate the average number of hundreds, fifties, and ducks (0 runs) by country.

```python
### 1. Find the average career length of Cricketers.
avg_career_Length = round(clear_data['Career_Length'].mean(),2)
print(f"The average career length of cricketers is {avg_career_Length} years.")
```

```python
### 2. Find the Average Batting Strike Rate for cricketers who played over 10 years.
avge_Batting_Strike_rate = round(clear_data[clear_data['Career_Length'] >10]['Batting_Strike_Rate'].mean(),2)
print(f"The average batting strike rate for cricketers who played over 10 years is {avge_Batting_Strike_rate}.")
```

```python
### 3. Find the number of cricketers who played before 1960.
number_crickets_before_1960 = len(clear_data[clear_data['Start_Date']<1960])
number_crickets_1960 = clear_data[clear_data['Start_Date']<1960].count()
print(f"number of crikects who played before 1960 is {number_crickets_before_1960}")
```

```python
### 4. Find max Highest Inns Score by Country.
max_Highest_Inns_Score = clear_data.groupby('Country')['Highest_Inns_Score'].max().to_frame().sort_values(by='Highest_Inns_Score',ascending=False)
print(max_Highest_Inns_Score)
```

```python
### 5. Hundreds, Fifties, Ducks (0) AVG by Country.
AVG = round(clear_data.groupby('Country')[['4s','6s','0']].mean(), 2).sort_values(by='4s',ascending=False)
print(AVG)
```

```python
### Top 10 Players by Highest Innings Score.
top_10_players = clear_data.groupby('Player')['Highest_Inns_Score'].sum().nlargest(10).reset_index()
plt.figure(figsize=(10, 6))
plt.rcParams.update({
    'figure.facecolor': '#1b263c',  # figure background
    'axes.facecolor': '#1b263c',
    'axes.edgecolor': 'white',      # spine color
    'axes.labelcolor': 'white',     # xlabel and ylabel color
    'xtick.color': 'white',         # x-axis tick color
    'ytick.color': 'white',         # y-axis tick
    'text.color': 'white',          # general text (title, etc.)
    'axes.titleweight': 'bold',     # bold title
    'axes.labelweight': 'bold',     # bold labels
})
sns.barplot(data=top_10_players, x='Player', y='Highest_Inns_Score', color='green')
plt.xlabel('Player')
plt.ylabel('Highest Innings Score')
plt.title('Top 10 Players by Highest Innings Score')
plt.xticks(rotation=90)
```

```python
### 4. Number of Hundreds, Fifties, and Ducks by Country
Numbers_100_50_0 = clear_data.groupby('Country')[['100', '50', '0']].sum()
Numbers_100_50_0
```

```python
plt.figure(figsize=(10, 6))
plt.rcParams.update({
    'figure.facecolor': '#1b263c',  # figure background
    'axes.facecolor': '#1b263c',
    'axes.edgecolor': 'white',      # spine color
    'axes.labelcolor': 'white',     # xlabel and ylabel color
    'xtick.color': 'white',         # x-axis tick color
    'ytick.color': 'white',         # y-axis tick
    'text.color': 'white',          # general text (title, etc.)
    'axes.titleweight': 'bold',     # bold title
    'axes.labelweight': 'bold',     # bold labels
})
ax = Numbers_100_50_0.plot(kind = 'barh', width = 0.8, stacked = True, cmap = 'mako', figsize = (12,6))
for i, (country, row) in enumerate(Numbers_100_50_0.iterrows()):
    values = row.tolist()
    label = f"100s: {values[0]:,}, 50s: {values[1]:,}, 0s: {values[2]:,}"

    total = sum(values)
    ax.text(
        total + 10,
        i,
        label,
        va = 'center',
        fontsize = 9,
    )


plt.title('Hundreds, Fifties, and Ducks by Country', fontsize = 14)
plt.ylabel('Country', fontsize = 12)
plt.xlabel('Scores', fontsize = 12)
plt.xlim(0,1200)

plt.tight_layout();
plt.legend(bbox_to_anchor = (1.09, 0.9));
```
